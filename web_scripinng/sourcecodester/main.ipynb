```python
class Article(object):
    def __init__(self):
        self._name = None
        self._link = None
        self._download = None
        self._title = None
        self._description = None
        self._views = None
        self._img = None
        self._post_info =None
        self._post_by = None
    # Name
    @property
    def name(self):
        return self.name

    @name.setter
    def name(self,name):
        self._name = name
    # Link
    @property
    def link(self):
        return self._link

    @link.setter
    def link(self,link):
        self._link = link
    # Download
    @property
    def download(self):
        return self._download

    @download.setter
    def download(self,download):
        self._download = download
    
    # title
    @property
    def title(self):
        return self._title

    @title.setter
    def title(self,title):
        self._title = title

    # description
    @property
    def description(self):
        return self._description

    @description.setter
    def description(self,description):
        self._description = description

    # views
    @property
    def views(self):
        return self._views

    @views.setter
    def views(self,views):
        self._views = views

    # img
    @property
    def img(self):
        return self._img

    @img.setter
    def img(self,img):
        self._img = img

    # post_info
    @property
    def post_info(self):
        return self._post_info

    @post_info.setter
    def post_info(self,post_info):
        self._post_info = post_info

    # post_by
    @property
    def post_by(self):
        return self._post_by

    @post_by.setter
    def post_by(self,post_by):
        self._post_by = post_by

    def getObj(self):
        return {
            "name":self._name,
            "link":self._link,
            "download":self._download, 
            "title":self._title, 
            "description":self._description, 
            "views":self._views, 
            "img":self._img, 
            "post_info":self._post_info,
            "post_by":self._post_by 
        }
```

```python
async def callApi(page):
    pages = []
    no=page
    page = requests.get(urlPage+page)
    soup = BeautifulSoup(page.content, 'html.parser')
    results = soup.find(class_='view-content')
    _elems = results.find_all('article', class_='node')
    for _elem in _elems:
        if not None in (_elem):
            obj = Article()
            obj.name = _elem.find(class_="field--name-title").text
            obj.post_info = _elem.find('span',class_="post-info").text
            obj.post_by = _elem.find(class_='username').text
            # print(_elem.find('h2',class_="node__title").find('a')['href'])
            obj.link = _elem.find('h2',class_="node__title").find('a')['href']
            # print(url+"download-code?nid="+_elem['data-history-node-id'])
            obj.download = url+"download-code?nid="+_elem['data-history-node-id']
            # print(_elem.find('li',class_="statistics-counter").text)
            obj.views = _elem.find('li',class_="statistics-counter").text
            obj.title = _elem['data-history-node-id']
            # image-style-thumbnail
            obj.img = url + _elem.find('img',class_="image-style-thumbnail")['src']
            # print(_elem.find('div',class_="field--type-text-with-summary").text)
            obj.description = _elem.find('div',class_="field--type-text-with-summary").text
            # field--type-text-with-summary
            # https://www.sourcecodester.com/download-code?nid=12062&title=Android+Simple+Quiz+App+Tutorial+with+Source+Code
            # statistics-counter
            # print(_elem.find(class_="field--name-title").text)
            # print(obj.post_info)
            # print(obj.post_by)
            # pages.append(json.dumps(vars(obj),indent=4))
            pages.append(obj.getObj())
            # jdata[page].append(obj.name)
            # print(obj.getObj())
    
    jdata[no]=pages
    # print(pages)
    # print(len(jdata[0]))
    # saveJsonFile()
```
