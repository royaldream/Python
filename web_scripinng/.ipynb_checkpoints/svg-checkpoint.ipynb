```python
async def call_download(delay, data):
    await asyncio.sleep(delay)
    imgName=os.path.basename(data['img'].path)
    videoName = os.path.basename(data['video'].path)
    path = print(os.path.splitext(imgName)[0])
    try: 
        os.mkdir(path) 
    except OSError as error: 
        print(error)  
```

```python
async def callVideoList(videosList):
    for item in videosList:
        img_url = item['src']
        video_url = item['data-src']
        a_img = urlparse(img_url)
        a_video = urlparse(video_url)
        print(a_img.path,a_video.path)
        await asyncio.create_task(call_download(5000,{'img':a_img.path,'video':a_video.path}))
         
```

```python
asyncio.run(callVideoList(videosList))
```
